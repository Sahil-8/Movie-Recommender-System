```python
movies = pd.read_csv('/content/tmdb_5000_movies.csv', encoding='latin-1') # Load movies data, update path and specify encoding
credits = pd.read_csv('/content/tmdb_5000_credits.csv', encoding='latin-1') # Load credits data, update path and specify encoding
movies = movies.merge(credits, on='title')
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
# Ensure the following code blocks are executed before running new['tags'][0]:
import pandas as pd
import ast

# Convert 'genres', 'keywords', 'cast', and 'crew' to lists before concatenation:
movies['genres'] = movies['genres'].apply(lambda x: x if isinstance(x, list) else [x])
movies['keywords'] = movies['keywords'].apply(lambda x: x if isinstance(x, list) else [x])
# The 'cast' and 'crew' columns were previously dropped,
# so you need to create these columns again before using them
movies['cast'] = movies['cast'] if 'cast' in movies else ''  # Add this line
movies['cast'] = movies['cast'].apply(lambda x: x if isinstance(x, list) else [x])
movies['crew'] = movies['crew'] if 'crew' in movies else ''  # Add this line
movies['crew'] = movies['crew'].apply(lambda x: x if isinstance(x, list) else [x])

movies['overview'] = movies['overview'].astype(str).apply(lambda x:x.split())
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new['tags'] = new['tags'].apply(lambda x: " ".join(x))

print(new['tags'][0])
```

```python
def recommend(movie):
  movie_index=new[new['title']==movie].index[0]
  distances=similarity[movie_index]
  movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movie_list: # Loop through movie_list
    print(new.iloc[i[0]]['title']) # Access the 'title' column using the index
```
